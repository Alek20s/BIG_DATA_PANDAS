```python
import pandas as pd

#TASK 1. mport the .csv file as a dataframe
df = pd.read_csv('HollywoodsMostProfitableStories.csv')

#ROUNDING numbers
df['Profitability'] = df['Profitability'].round(2)
df['Worldwide.Gross'] = df['Worldwide.Gross'].round(2)

# Droping rows with missing values and assigning df to new data frame df1
df1 = df.dropna()

# Dropping duplicates
df2 = df1.drop_duplicates()

# conditional removing rows
df2 = df2[df2['Profitability'] > 0.3]
df2 = df2[df2['Profitability'] < 14.0]
df2 = df2[df2['Worldwide.Gross'] > 8.5]
df2 = df2[df2['Worldwide.Gross'] < 500]



print(df2)
print(df2.shape)

#df2.to_csv('df_cleaned.csv')



```
